```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import sobel_h, sobel_v
```

```python
stack5 = io.imread('Stack5.tif')
stack9 = io.imread('Stack9.tif')
```

```python
for i, image in enumerate(stack5):
    plt.imshow(image, cmap="gray")
    plt.show()
```

```python
der = np.gradient(stack5)
var = np.var(der)
plt.hist(var,bins = 100, color="blue",alpha=0.5)
```

```python
der = np.gradient(stack9)
var = np.var(der)
plt.hist(var,bins = 100, color="blue",alpha=0.5)
```

```python
deriv_h = [np.gradient(image, axis=1) for image in stack5]
deriv_v = [np.gradient(image, axis=0) for image in stack5]

variances1 = [np.var(deriv) for deriv in zip(deriv_h, deriv_v)]

plt.boxplot(variances1)
plt.title("Variances des drives")
plt.show()
```

```python
deriv_h = [np.gradient(image, axis=1) for image in stack9]
deriv_v = [np.gradient(image, axis=0) for image in stack9]

variances2 = [np.var(deriv) for deriv in zip(deriv_h, deriv_v)]

plt.boxplot(variances2)
plt.title("Variances des drives")
plt.show()
```

```python


for i, image in enumerate(stack5):
    spec = np.fft.fft2(image)
    spec = np.fft.fftshift(spec)
    mag = np.abs(spec)

    plt.imshow(np.log(mag + 1), cmap='gray')

    plt.title(f"Spectre de l'image 5{i + 1}")

    plt.show()
```

```python

for i, image in enumerate(stack9):
    spec = np.fft.fft2(image)
    spec = np.fft.fftshift(spec)
    mag = np.abs(spec)

    plt.imshow(np.log(mag + 1), cmap='gray')

    plt.title(f"Spectre de l'image 9{i + 1}")

    plt.show()
```
